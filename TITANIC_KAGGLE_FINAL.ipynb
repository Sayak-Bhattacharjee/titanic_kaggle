```python
def male_female_child(passenger):
    age,sex=passenger
    
    if age<14:
        return "child"
    else:
        return sex
```

```python
def gender_determination(char):
    if char == 'female' or char == 'child':
        return 1
    else:
        return 0
x_titanic_df["Gender"]=titanic_df["Sex"].apply(gender_determination)
```

```python
def singleornot(family):
    sib,parch=family
    if sib==0 and parch==0:
        return "SINGLE"
    else:
        return "With Family"
titanic_df["Single Or Not"]=titanic_df[["SibSp","Parch"]].apply(singleornot,axis=1)
```
